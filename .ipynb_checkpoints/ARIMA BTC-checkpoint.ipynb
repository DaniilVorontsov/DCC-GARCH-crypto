```python
import BTCUSD_hist_ByBit as bbyt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
btc_hist=bbyt.get_historic_data(date='2022-10-31',years=2)
```

```python
btc_hist.head()
```

```python
btc_hist.shape
```

```python
btc_hist['datetime']=btc_hist.datetime.dt.strftime('%Y-%m-%d')
```

```python
df=btc_hist.groupby('datetime',as_index=False).agg({'close':'mean'})
```

```python
fig,ax=plt.subplots(figsize=(12,8))
ax.plot(df.datetime,df.close)
ax.grid()
ax.xaxis.set_major_locator(plt.MaxNLocator(10))
ax.set_title('BTC, USD')
plt.show()
```

```python
df['pct_change']=df.close.pct_change().fillna(0)
```

```python
fig,ax=plt.subplots(figsize=(12,8))
ax.plot(df.datetime,df['pct_change'])
ax.grid()
ax.xaxis.set_major_locator(plt.MaxNLocator(10))
ax.set_title('pct change(d), USD')
plt.show()
```

```python
df.close.shift(1).fillna(0)
```

```python
df['abs_change']=(df.close.shift(1)-df.close).fillna(0)
```

```python
fig,ax=plt.subplots(figsize=(12,8))
ax.plot(df.datetime,df['abs_change'])
ax.grid()
ax.xaxis.set_major_locator(plt.MaxNLocator(10))
ax.set_title('abs change(d), USD')
plt.show()
```
