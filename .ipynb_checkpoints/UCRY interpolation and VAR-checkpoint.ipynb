```python
import pandas as pd
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt

from statsmodels.tsa.api import adfuller
from statsmodels.tsa.vector_ar.var_model import VAR
import statsmodels.api as sm
```

```python
UCRY=pd.read_csv('UCRY Weekly Data Updating_Available until 2022 Q4.csv')

UCRY['date']=pd.to_datetime(UCRY.Timeline.str.split('-', expand=True)[1], dayfirst=True)

UCRY=UCRY.set_index('date').drop(columns=['Timeline','Date'])
```

```python
UCRY.plot(figsize=(12,8))
plt.grid()
plt.title('UCRY Indexes')
plt.show()
```

```python
(UCRY['UCRY Policy Index']-UCRY['UCRY Price Index']).plot(figsize=(12,8))
plt.title('Разность индексов, %')
plt.grid()
plt.show()
```

```python
x = np.arange(len(UCRY))
xs=pd.date_range(start='2014-01-05',end='2023-01-01')
UCRY_daily=pd.DataFrame(xs).rename(columns={0:'date'}).set_index('date')\
    .join(UCRY,how='left')\
    .interpolate(method='time')

#UCRY_daily.to_csv('UCRY_daily.csv')
```

```python
coins=pd.read_csv('daily_Rt.csv',index_col=0)
```

```python
coins.index=coins.index.astype('datetime64[ns]')
```

```python
df=coins.dropna().join(UCRY_daily).dropna()
```

```python
df.to_csv('./скрипты R/UCRY_daily_short.csv')
```
