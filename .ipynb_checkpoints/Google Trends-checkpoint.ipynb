```python
import pandas as pd
from pytrends.request import TrendReq
from time import sleep
import datetime as dt
```

```python
pytrends=TrendReq()
kw_list1= ['bitcoin','Mining','Tesla','Blockchain','Elon Musk']
pytrends.build_payload(kw_list1, cat=0, timeframe='2018-01-01 2023-01-31')
df1=pytrends.interest_over_time()
```

```python
pytrends=TrendReq()
kw_list2=['Coin','Cryptocurrency','Ethereum',]
pytrends.build_payload(kw_list2, cat=0, timeframe='2018-01-01 2023-01-31')
df2=pytrends.interest_over_time()
```

```python
df=pd.merge(df1.drop('isPartial',axis=1), df2.drop('isPartial',axis=1), left_index=True, right_index=True)
```

```python
index=pd.Series(pd.date_range(start='2018-01-01',end='2023-01-31'),name='date')
```

```python
df_fin=pd.merge(pd.DataFrame(index),df,how='left', on='date')
```

```python
df_fin[:10]
```

```python
df_fin=df_fin.fillna(method='backfill').fillna(method='ffill')
```

```python
df_fin.set_index('date',inplace=True)
```

```python
df_fin.to_csv('Google_trends_data.csv')
```
